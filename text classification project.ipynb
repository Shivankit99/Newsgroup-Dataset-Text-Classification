```python

number_of_docs = 0
dictionary = {}
for folder in subfolders:
    subfolder_path = original_path + '\\' + folder
    docs = os.listdir(subfolder_path)
    for doc in docs:
        file_path = subfolder_path + '\\' + doc
        handle = open(file_path, 'r', encoding = 'utf-8', errors="surrogateescape")
        number_of_docs+=1
        lines = handle.readlines()
        for line in lines:
            string = line.split(" ")
            for i in string:
                if i not in stop_words and i not in dictionary:
                    dictionary[i] = 0            
                if i in dictionary:
                    dictionary[i] +=1
```

```python
def create_count_dictionary(x_train,y_train,feature_list):
    result = {}
    all_classes = set(y_train)
    total_count = 0
    for curr_class in all_classes:
        result[curr_class] = {}
        current_rows = y_train == curr_class
        curr_x = x_train[current_rows]
        result[curr_class]["Total_Files"] = curr_x.shape[0]
        for word in feature_list:
            index = feature_list.index(word)
            count = curr_x[:,index].sum()
            total_count += count
            result[curr_class][word] = count
    result["Total_Count"] = len(y_train)
    return result
```

```python
count_dictionary = create_count_dictionary(x_train,y_train,feature_list)
```

```python
def probability(dictionary, x, current_class,feature_list):
    output = np.log(dictionary[current_class]["Total_Files"]) - np.log(dictionary["Total_Count"])
    for j in range(len(feature_list)):
        if dictionary[current_class] == 'Total_Files' :
            continue
        if(x[j] == 0):
            current_xj_probablity = 0
        else:
            count_current_class_with_value_xj = dictionary[current_class][feature_list[j]] + 1
            count_current_class = dictionary[current_class]["Total_Files"] + len(dictionary[current_class].keys())
            current_xj_probablity = np.log(count_current_class_with_value_xj) - np.log(count_current_class)
        output = output + current_xj_probablity
    return output
```

```python
def predictSinglePoint(dictionary, x,feature_list):
    classes = dictionary.keys()
    best_p = -1000
    best_class = -1
    first_run = True
    for current_class in classes:
        if current_class == "Total_Count":
            continue
        p_current_class = probability(dictionary, x, current_class,feature_list)
        if (first_run or p_current_class > best_p):
            best_p = p_current_class
            best_class = current_class
        first_run = False
    return best_class
```

```python
def predict(dictionary, X_test,feature_list):
    y_pred = []
    for x in X_test:
        x_class = predictSinglePoint(dictionary, x,feature_list)
        y_pred.append(x_class)
    return y_pred
```
